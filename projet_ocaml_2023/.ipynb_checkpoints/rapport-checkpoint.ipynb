```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# read excel
df_ods = pd.read_excel('test_stats.ods', engine='odf')
```

```python
# mesure liste time for the compression of (2 pow n) node
sns.lineplot(x=df_ods['N'], y=df_ods['time_liste(us)'])

plt.title('compression_time_liste(us)')
plt.xlabel('N')
plt.ylabel('time')
plt.show()
```

```python
# mesure arbre time for the compression of (2 pow n) node
sns.lineplot(x=df_ods['N'], y=df_ods['time_arbre(us)'])

plt.title('compression_time_arbre(us)')
plt.xlabel('N')
plt.ylabel('time')
plt.show()
```

```python
# mesure liste space for the compression of (2 pow n) node

sns.lineplot(x=df_ods['N'], y=df_ods['space_liste(mots)'])

plt.title('compression_space_liste')
plt.xlabel('N')
plt.ylabel('space')
plt.show()
```

```python
# mesure arbre space for the compression of (2 pow n) node

sns.lineplot(x=df_ods['N'], y=df_ods['space_arbre(mots)'])

plt.title('compression_space_arbre')
plt.xlabel('N')
plt.ylabel('space')
plt.show()
```

```python
# mesure taux for the compression of (2 pow n) node

sns.lineplot(x=df_ods['N'], y=df_ods['taux(%)'])

plt.title('compression_taux')
plt.xlabel('N')
plt.ylabel('taux(%)')
plt.show()
```
